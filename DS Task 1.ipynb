## A. Student Scores Tracker

```python
#1.Creating a list of scores

Test_scores = [65, 78, 90, 55, 88, 72]
```

```python
#2.The highest score

Highest_score = max(Test_scores) 

print(Highest_score)
```

```python
#3.The lowest score

Lowest_score = min(Test_scores) 

print(Lowest_score)
```

```python
 #4.The average score

Average_score = sum(Test_scores)/len(Test_scores)
```

```python
print(Average_score)
```

```python
#5.Updating the list using append

Test_scores.append(80)
```

```python
#6.Removing the failed score from the list

#Assuming the failing score < 60

Test_scores.remove(55)
```

```python
#7.Final Test_scores list

print(Test_scores)
```
